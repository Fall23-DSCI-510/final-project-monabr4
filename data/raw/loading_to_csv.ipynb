```python
import requests
from datetime import datetime, timedelta
import pandas as pd
```

 Extracted the population for each state in Malaysia. 

```python
url = "https://myvaccination-backend.vercel.app/api/pop"
headers = {}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    data = response.json()

    clean_data = []
    for entry in data:
        state_name = entry.get('stateName')
        total_population = entry.get('total')

        clean_data.append({'stateName': state_name, 'total': total_population})
    population = pd.DataFrame(clean_data)
    population.to_csv('cleaned_population_data.csv', index=False)
else:
    print(f"Error: {response.status_code}")
    print(response.text)

    #manually changed "stateName to name" in the static file to match the geojson file.
```

Exporting Johor Data 

```python
base_url = "https://covid-19.samsam123.name.my/api/state"
start_date_str = "2020-01-25"
end_date_str = "2023-01-25" #None if wanted real-time update
state_name = "Johor"

start_date = datetime.strptime(start_date_str, "%Y-%m-%d")
end_date = datetime.strptime(end_date_str, "%Y-%m-%d")

current_date = start_date

date_list = []
state_list = []
cases_new_list = []
cases_recovered_list = []
cases_active_list = []
cases_child_list = []
cases_adolescent_list = []
cases_adult_list = []
cases_elderly_list = []

while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")
    url = f"{base_url}?date={date_str}&state={state_name}"
    response = requests.get(url)

    if response.status_code == 200:
        covid_data = response.json()

        date_list.append(date_str)
        state_list.append(covid_data['state'])
        cases_new_list.append(int(covid_data['cases_new']))
        cases_recovered_list.append(int(covid_data['cases_recovered']))
        cases_active_list.append(int(covid_data['cases_active']))
        cases_child_list.append(int(covid_data['cases_child']))
        cases_adolescent_list.append(int(covid_data['cases_adolescent']))
        cases_adult_list.append(int(covid_data['cases_adult']))
        cases_elderly_list.append(int(covid_data['cases_elderly']))
    else:
        print(f"Failed to fetch data for {date_str}")

    current_date += timedelta(days=1)
#Variables of focus 
data = {
    'Date': date_list,
    'state': state_list,
    'cases_new': cases_new_list,
    'cases_recovered': cases_recovered_list,
    'cases_active': cases_active_list,
    'cases_child': cases_child_list,
    'cases_adolescent': cases_adolescent_list,
    'cases_adult': cases_adult_list,
    'cases_elderly': cases_elderly_list,
}

johor = pd.DataFrame(data)

# Converting it to CSV
csv_filename = 'Johor.csv' 
johor.to_csv(csv_filename, index=False)
```
