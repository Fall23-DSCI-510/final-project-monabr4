```python
import pandas as pd
```

```python
#States 
df1 = pd.read_csv('Johor.csv')
df2 = pd.read_csv('Kedah.csv')
df3 = pd.read_csv('Kelantan.csv')
df4 = pd.read_csv('Melaka.csv')
df5 = pd.read_csv('Negeri_Sembilan.csv')
df6 = pd.read_csv('Pahang.csv')
df7 = pd.read_csv('Perak.csv')
df8 = pd.read_csv('Perlis.csv')
df9 = pd.read_csv('Pulau_Pinang.csv')
df10 = pd.read_csv('Sabah.csv')
df11= pd.read_csv('Sarawak.csv')
df12 = pd.read_csv('Terengganu.csv')
df13= pd.read_csv('Selangor.csv')
df14= pd.read_csv('W_P_Kuala_Lumpur.csv')
df15= pd.read_csv('W_P_Putrajaya.csv')


merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15], axis=0, ignore_index=True)
merged_df.to_csv('state_covid_data.csv', index=False)
```

```python
# Merging states and population
covid_data = pd.read_csv('state_covid_data.csv')
population_data = pd.read_csv('/Users/mona/final-project-monabr4/data/raw/exported_statis_csv_shapefile/cleaned_population_data.csv')
merged_data = pd.merge(covid_data, population_data, on='state', how='left')
merged_data = merged_data.rename(columns={'total': 'population'})
merged_data['cases_new_per_capita'] = merged_data['cases_new'] / merged_data['population']
merged_data['cases_recovered_per_capita'] = merged_data['cases_recovered'] / merged_data['population']
merged_data['cases_active_per_capita'] = merged_data['cases_active'] / merged_data['population']
merged_data.to_csv('/Users/mona/final-project-monabr4/data/final_covid_data.csv', index=False)
```
